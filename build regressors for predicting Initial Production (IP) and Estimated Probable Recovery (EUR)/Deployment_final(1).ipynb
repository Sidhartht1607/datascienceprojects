```python
# importing libraries
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.impute import KNNImputer
import warnings
warnings.filterwarnings('ignore')
```

```python
#importing Shale_HO deployment dataset
```

```python
data = pd.read_csv("Shale_HO.csv")
data
```

```python
#changing datetime
```

```python
data['FirstProdDate'] = pd.to_datetime(data['FirstProdDate'],format='%Y').dt.year
data.describe()
```

```python
#replacing corrupted values with np.nan
```

```python
data.replace([np.inf, -np.inf, -999, '', 'inf', 'NaaN', 'NaN', None,' df ', 'XXX'], np.nan, inplace=True)
```

```python
data.isna().sum()
```

```python
#using the saved encoders and imputer
```

```python
import pickle 

le_load2 = pickle.load( open( 'ShaleGas_encoder2.sav', "rb" ) )
data['FluidType'] = le_load2.transform(data['FluidType'].astype(str))
```

```python
le_load1 = pickle.load( open( 'ShaleGas_encoder1.sav', "rb" ) )
data['Operator'] = le_load1.transform(data['Operator'].astype(str))
```

```python
imp_load = pickle.load( open( 'ShaleGas_imp.sav', "rb" ) )

data_imputed = imp_load.transform(data)
```

```python
data_imputed = pd.DataFrame(data_imputed)
data_imputed.columns = data.columns
```

```python
data_imputed['Operator'] = data_imputed['Operator'].astype(int)
data_imputed['FluidType']= data_imputed['FluidType'].astype(int)
```

```python
model_outlier_load = pickle.load( open( 'ShaleGas_outlier.sav', "rb" ) )#outlier model\
anomaly25 = model_outlier_load.predict(data_imputed)
data_imputed = data_imputed[anomaly25==1]
```

```python
## Popping the output 
```

```python
EUR=data_imputed.pop('EUR')
data_imputed
```

```python
#creating new features and other datapreprocessing done on the training dataset.
```

```python
X1 = data_imputed
```

```python
X1['TotalProppant'] = np.log(X1['TotalProppant'])
X1['TotalFluid'] = np.log(X1['TotalFluid'])
X1['TotalProppant'] = np.log(X1['TotalProppant'])
X1['TotalFluid'] = np.log(X1['TotalFluid'])
```

```python
X1['Fracture Length'] = X1['BottomPerf'] - X1['TopPerf']
X1['Net Pay'] = X1['Isopach'] * X1['Porosity']
X1['Initial Permeability'] = 0.001 * X1['FlowGas']/(X1['Pressure']*X1['LateralLength']*0.5*(X1['SurfaceX']+X1['SurfaceY']))
```

```python
X1.shape
```

```python
#Using RobustScaler
```

```python
model_scaler = pickle.load( open( 'Shale_Gas_RobustScaler.sav', "rb" ) )

abcd= model_scaler.transform(X1[[ 'TD', 'TVD', 'TotalFluid', 'TotalProppant', 'TopPerf',
       'BottomPerf',  'LateralLength', 'ChokeTopDescription',
       'MaxFlowingPSI', 'MaxShutinPSI', 'FlowGas', 'FlowWater', 'SurfaceX',
       'SurfaceY', 'Z_BelowSurface', 'Isopach', 'Porosity',
       'TotalGasBCFPerSection', 'Temperature', 'Pressure', 'KOP',
       'FirstProdDate', 'minLateralSpacing', 'numOfOffsetLateralSpacingWells',
       'Fracture Length', 'Net Pay', 'Initial Permeability','IP']])

X1[[ 'TD', 'TVD', 'TotalFluid', 'TotalProppant', 'TopPerf',
       'BottomPerf',  'LateralLength', 'ChokeTopDescription',
       'MaxFlowingPSI', 'MaxShutinPSI', 'FlowGas', 'FlowWater', 'SurfaceX',
       'SurfaceY', 'Z_BelowSurface', 'Isopach', 'Porosity',
       'TotalGasBCFPerSection', 'Temperature', 'Pressure', 'KOP',
       'FirstProdDate', 'minLateralSpacing', 'numOfOffsetLateralSpacingWells',
       'Fracture Length', 'Net Pay', 'Initial Permeability','IP']]=abcd
```

```python
#removing the columns based on F_score and Mutual Information Regression
```

```python
model_f_score = pickle.load( open( 'ShaleGas_fscore.sav', "rb" ) )
model_f_score.shape
```

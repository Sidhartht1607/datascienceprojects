```python
Features.describe()
```

```python
from sklearn.preprocessing import RobustScaler
from sklearn.ensemble import IsolationForest
```

```python
robust=RobustScaler() 
robust_features=robust.fit_transform(Features)
s_Features=pd.DataFrame(robust_features)
s_Features.columns=['PHIND','DeltaPHI','PE','NM_M','RELPOS','GR','ILD_log10']
```

```python
from sklearn import preprocessing
model = IsolationForest(contamination = 0.2)
model.fit(s_Features)
s_Features['anomaly25']=pd.Series(model.predict(s_Features))
print(s_Features)
s_Features['Facies_Label']=df['Facies_Label']
print((s_Features['anomaly25']).value_counts())
s_Features.loc[s_Features['anomaly25']==1]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(s_Features[['PHIND','DeltaPHI','PE','NM_M','RELPOS','GR','ILD_log10']],s_Features['Facies_Label'],test_size=0.4,stratify=s_Features['Facies_Label'])
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(penalty='l2', C=4714, max_iter=853, solver='sag')
```

```python
classifier.fit(X_train,y_train)
```

```python
print(classifier.coef_) #a1 and a2
print(classifier.intercept_)
```

```python
prediction = classifier.predict(X_test)
```

```python
print(prediction)
print(y_test)
print(classifier.score(X_test,y_test)) ## generalization score
print(classifier.score(X_train,y_train))
```

```python
from sklearn.metrics import f1_score
print(f1_score(y_test,classifier.predict(X_test), average ='weighted'))
print(f1_score(y_train,classifier.predict(X_train), average='weighted'))
```

```python
from sklearn.metrics import precision_score
print(precision_score(y_test,classifier.predict(X_test),average='weighted')) ## generalization score
print(precision_score(y_train,classifier.predict(X_train),average='weighted'))
```

```python
from sklearn.metrics import recall_score
print(recall_score(y_test,classifier.predict(X_test),average='weighted')) ## generalization score
print(recall_score(y_train,classifier.predict(X_train),average='weighted'))
```

```python
    from sklearn import linear_model, decomposition, datasets
    from sklearn.pipeline import Pipeline
    from sklearn.model_selection import GridSearchCV
```

```python
pca = decomposition.PCA()
logistic_Reg = linear_model.LogisticRegression()
pipe = Pipeline(steps=[('pca', pca),
                           ('logistic_Reg', logistic_Reg)])
```

```python
n_components = list(range(1,X_test.shape[1]+1,1))

C = np.logspace(-4, 4, 50)
penalty = ['l1', 'l2']
solver=['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']


parameters = dict(pca__n_components=n_components,
                logistic_Reg__C=C,
                logistic_Reg__penalty=penalty,
                log_solver=solver)

clf_GS = GridSearchCV(pipe, parameters)
clf_GS.fit(X_train, y_train)

print('Best Penalty:', clf_GS.best_estimator_.get_params()['logistic_Reg__penalty'])
print('Best C:', clf_GS.best_estimator_.get_params()['logistic_Reg__C'])
print('Best Number Of Components:', clf_GS.best_estimator_.get_params()['pca__n_components'])
print('Best Solver : ',clf_GS.best_estimator_.get_params()['log_solver'])
print(); print(clf_GS.best_estimator_.get_params()['logistic_Reg'])
```
